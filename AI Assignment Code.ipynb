## Prompt 2: "load necessary libraries and data for eda"

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Load the dataset
# Make sure the CSV file is downloaded and the path is correct
file_path = "/Users/joshbolton/Downloads/card_transdata.csv"  # replace with your local path if different
df = pd.read_csv(file_path)

# Display basic information about the dataset
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())
```

# Prompt 3: "describe the data, check the missing values and then drop them"

```python
# --- 1. Descriptive statistics ---
print("\nðŸ”¹ Statistical Summary:")
print(df.describe().T)

# --- 2. Check for missing values ---
print("\nðŸ”¹ Missing Values per Column:")
print(df.isnull().sum())

# --- 3. Drop rows with missing values (if any) ---
df = df.dropna()

# Confirm removal
print("\nâœ… After dropping missing values:")
print("New Shape:", df.shape)
print("Any missing left?", df.isnull().sum().sum() == 0)
```

## Univariate Analysis


# Prompt 4: "visualize fraud, used pin, repeat retail, online orders, ratio to median purchase price count distributions"

```python
# Set visualization style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# Visualize fraud distribution
plt.subplot(2, 2, 1)
sns.countplot(x='fraud', data=df, palette='coolwarm')
plt.title('Fraud Distribution')

# Visualize used_pin distribution
plt.subplot(2, 2, 2)
sns.countplot(x='used_pin_number', data=df, palette='crest')
plt.title('Used PIN Distribution')

# Visualize repeat_retailer distribution
plt.subplot(2, 2, 3)
sns.countplot(x='repeat_retailer', data=df, palette='mako')
plt.title('Repeat Retailer Distribution')

# Visualize online_order distribution
plt.subplot(2, 2, 4)
sns.countplot(x='online_order', data=df, palette='flare')
plt.title('Online Order Distribution')

plt.tight_layout()
plt.show()

# Visualize ratio_to_median_purchase_price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['ratio_to_median_purchase_price'], bins=50, kde=True, color='teal')
plt.title('Ratio to Median Purchase Price Distribution')
plt.xlabel('Ratio to Median Purchase Price')
plt.ylabel('Count')
plt.show()
```

## Bivariate Analysis: Transaction Type, Distance, and Price Ratio vs Fraud


# Prompt 5: "do bivariate distributions on transactions with pin/chip/online order and fraud transactions, percentage of fraud transactions with pin/chip/online order"

# Prompt 6: "graph percentage ones as well and put the fraudulent and non fradulent on the plots"

```python
# Bivariate relationships between transaction types and fraud
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.countplot(x='used_pin_number', hue='fraud', data=df, palette='coolwarm')
plt.title('Fraud vs Used PIN')

plt.subplot(1, 3, 2)
sns.countplot(x='used_chip', hue='fraud', data=df, palette='crest')
plt.title('Fraud vs Used Chip')

plt.subplot(1, 3, 3)
sns.countplot(x='online_order', hue='fraud', data=df, palette='mako')
plt.title('Fraud vs Online Order')

plt.tight_layout()
plt.show()

# Calculate percentage of fraud transactions for each category
fraud_pin = df.groupby('used_pin_number')['fraud'].mean() * 100
fraud_chip = df.groupby('used_chip')['fraud'].mean() * 100
fraud_online = df.groupby('online_order')['fraud'].mean() * 100

# Combine into one DataFrame for clarity
fraud_percentages = pd.DataFrame({
    'Used_PIN (%)': fraud_pin,
    'Used_Chip (%)': fraud_chip,
    'Online_Order (%)': fraud_online
})

print("ðŸ”¹ Percentage of Fraud Transactions by Transaction Type:")
print(fraud_percentages)
```

```python
# Fraud vs Non-Fraud percentage for Used PIN
pin_counts = df.groupby('used_pin_number')['fraud'].value_counts(normalize=True).unstack() * 100
pin_counts.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(6, 4))
plt.title('Fraud vs Non-Fraud by Used PIN')
plt.ylabel('Percentage (%)')
plt.xlabel('Used PIN')
plt.legend(['Non-Fraud', 'Fraud'])
plt.tight_layout()
plt.show()

# Fraud vs Non-Fraud percentage for Used Chip
chip_counts = df.groupby('used_chip')['fraud'].value_counts(normalize=True).unstack() * 100
chip_counts.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(6, 4))
plt.title('Fraud vs Non-Fraud by Used Chip')
plt.ylabel('Percentage (%)')
plt.xlabel('Used Chip')
plt.legend(['Non-Fraud', 'Fraud'])
plt.tight_layout()
plt.show()

# Fraud vs Non-Fraud percentage for Online Order
online_counts = df.groupby('online_order')['fraud'].value_counts(normalize=True).unstack() * 100
online_counts.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(6, 4))
plt.title('Fraud vs Non-Fraud by Online Order')
plt.ylabel('Percentage (%)')
plt.xlabel('Online Order')
plt.legend(['Non-Fraud', 'Fraud'])
plt.tight_layout()
plt.show()


```

# Prompt 7: "make 2 separate scatter plots with the distance columns vs the price ratio"

```python
# Scatter plot: distance_from_home vs ratio_to_median_purchase_price
plt.figure(figsize=(8, 5))
sns.scatterplot(x='distance_from_home', y='ratio_to_median_purchase_price', data=df, hue='fraud', palette='coolwarm', alpha=0.6)
plt.title('Distance from Home vs Ratio to Median Purchase Price')
plt.xlabel('Distance from Home')
plt.ylabel('Ratio to Median Purchase Price')
plt.legend(title='Fraud')
plt.tight_layout()
plt.show()

# Scatter plot: distance_from_last_transaction vs ratio_to_median_purchase_price
plt.figure(figsize=(8, 5))
sns.scatterplot(x='distance_from_last_transaction', y='ratio_to_median_purchase_price', data=df, hue='fraud', palette='coolwarm', alpha=0.6)
plt.title('Distance from Last Transaction vs Ratio to Median Purchase Price')
plt.xlabel('Distance from Last Transaction')
plt.ylabel('Ratio to Median Purchase Price')
plt.legend(title='Fraud')
plt.tight_layout()
plt.show()
```
